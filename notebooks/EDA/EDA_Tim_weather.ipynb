```python
import seaborn as sns
import matplotlib.pyplot as plt

from waterlevels_oker.data import preprocess_weather_data

%matplotlib inline
%config InlineBackend.figure_format = "retina"
plt.style.use("default")
```

```python
weather = preprocess_weather_data()
```

```python
weather.info()
```

```python
f"Percentage of columns with na: {round(weather.isna().any(axis=1).sum()/len(weather), 2)*100}"
```

```python
for col in weather.select_dtypes("number"):
    sns.histplot(weather[col])
    plt.show()
```

```python
for col in weather.columns:
    fig, ax = plt.subplots(figsize=(12,4))
    sns.lineplot(weather[col], ax=ax)
    plt.show()
```
