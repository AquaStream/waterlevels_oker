```python
import seaborn as sns
import matplotlib.pyplot as plt
from waterlevels_oker.data import preprocess_ohrum_data

%matplotlib inline
%config InlineBackend.figure_format = "retina"
plt.style.use("default")
```

```python
ohrum = preprocess_ohrum_data()
```

```python
ohrum
```

```python
ohrum.info()
```

```python
fig, ax = plt.subplots(figsize=(12,4))
sns.lineplot(ohrum["Waterlevel relative [cm]"], ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(12,4))
for year in range(2005, 2025):
    sns.lineplot(x=ohrum.loc[str(year)].index.dayofyear, y=ohrum.loc[str(year),"Waterlevel relative [cm]"], ax=ax)
ax.set_xlabel("")
ax.tick_params(
    axis='x',
    labelbottom=False)
ax.set_title("2005 â€“ 2024")
plt.show()
```
